```python
import os
from openai import OpenAI
from rich.console import Console
from rich.panel import Panel
from rich.columns import Columns
from rich.text import Text
from rich.table import Table
from IPython.display import display, Markdown

# ── Set your OpenRouter API key ──────────────────────────────────────────────
# Option 1: paste it here
os.environ["OPENROUTER_API_KEY"] = "API key"

# Option 2: export OPENROUTER_API_KEY in your shell before launching Jupyter

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.environ["OPENROUTER_API_KEY"],
)

console = Console()
print("✅ OpenRouter client ready")
```

---
## Model Definitions

```python
All_models = [
    "openai/gpt-4o",
    "openai/gpt-5.5", # hybrid
    "anthropic/claude-sonnet-4",
    "anthropic/claude-3-haiku",
    "meta-llama/llama-3-70b-instruct",
    "deepseek/deepseek-r1",
    "deepseek/deepseek-v4-flash"
    "x-ai/grok-4.3" ,
    "openai/o1"]


# Below models can be turn on/off reasoning mode
HYBRID_MODLES = [
    "deepseek/deepseek-r1",
    "openai/o1"
    "openai/gpt-5.5",
    "anthropic/claude-sonnet-4",
    "x-ai/grok-4.3",
    "deepseek/deepseek-v4-flash"
]

# Below models are non-reasoning models
NON_REASONING_MODELS = [
    "openai/gpt-4o",
    "anthropic/claude-3-haiku",
    "meta-llama/llama-3-70b-instruct"
]
```

---
## Core Call Functions

```python
def generate_completion(
    model_id: str, 
    messages: list, 
    disable_reasoning: bool = False,
    reasoning_effort: str = "medium"
) -> str:
    """
    Generates text completions via OpenRouter, dynamically adjusting 
    reasoning parameters based on the model's structural capabilities.
    """

    
    extra_body_payload = {}

    # 2. Build the appropriate reasoning config payload
    if model_id in PURE_REASONING_MODELS:
        if disable_reasoning:
            # We can't stop the engine, but we can instruct OpenRouter to strip the text
            extra_body_payload["reasoning"] = {"exclude": True}
            
    elif model_id not in HYBRID_MODLES:
        # This handles Hybrid Models (Claude 4, GPT-5.5, Grok 4.3, Opus 4.7)
        if disable_reasoning:
            extra_body_payload["reasoning"] = {"effort": "none"}
        else:
            extra_body_payload["reasoning"] = {"effort": reasoning_effort}

    # 3. Execute the API request safely
    try:
        response = client.chat.completions.create(
            model=model_id,
            messages=messages,
            extra_body=extra_body_payload if extra_body_payload else None
        )
        return response.choices[0].message.content
        
    except Exception as e:
        print(f"OpenRouter API Error: {e}")
        return ""

```

## Demo 2 — Math Problem <a id="demo-02"></a>

[Go to Top](#Reasoning-LLMs)

```python
content = """
A store sells apples for $0.75 each and oranges for $1.10 each.
Sarah buys some apples and oranges and pays exactly $10.00 total.
She buys at least one of each fruit.

What are ALL possible combinations she could have bought?

Wrap math equations with $$ or $ in the output following LaTex Format
"""
prompt = [{"role": "user", "content": content}]
```

```python
## Tier 1 - deepseek-r1
display(
    Markdown(
        generate_completion(
        model_id="deepseek/deepseek-r1", 
        messages=prompt
        )
    )
)
```

--------
***converted below***

Sarah buys apples for $0.75 each and oranges for $1.10 each, paying exactly $10.00 while purchasing at least one of each fruit. The equation representing the total cost is:

0.75a + 1.10o = 10.00

where a is the number of apples and o is the number of oranges, both positive integers.

Converting to cents to eliminate decimals:

75a + 110o = 1000

Simplifying by dividing by 5:

15a + 22o = 200

Solving for integer solutions, the only pair (a, o) that satisfies the equation with a ≥ 1 and o ≥ 1 is a = 6, o = 5.

Verification:

* Cost of apples: 6 × 0.75 = 4.50
* Cost of oranges: 5 × 1.10 = 5.50
* Total: 4.50 + 5.50 = 10.00

No other combinations satisfy the conditions. Therefore, the only possible combination is 6 apples and 5 oranges.

All possible combinations:

* 6 apples and 5 oranges

Final Answer: (a, o) = (6, 5)

```python
## Tier 3 - deepseek-v4-flash
display(
    Markdown(
        generate_completion(
        model_id="deepseek/deepseek-v4-flash", 
        messages=prompt
        )
    )
)
```

----------
***convert below***
We need solutions to:

0.75a + 1.10o = 10.00

with integers a ≥ 1, o ≥ 1.

First multiply through by 100 to avoid decimals:

75a + 110o = 1000

Divide by 5:

15a + 22o = 200

Now solve for a:

15a = 200 - 22o

a = (200 - 22o) / 15

We require a to be a positive integer.

Compute modulo 15 for 200 - 22o ≡ 0 (mod 15):

* 200 mod 15 = 5 (since 15 × 13 = 195)
* 22 mod 15 = 7

So condition:

5 - 7o ≡ 0 (mod 15)

-7o ≡ -5 (mod 15)

7o ≡ 5 (mod 15)

Multiply both sides by the inverse of 7 mod 15. Since 7 × 13 = 91 ≡ 1 (mod 15), inverse is 13:

o ≡ 5 × 13 = 65 ≡ 5 (mod 15)

So o = 5 + 15k.

Try small k:

* k = 0: o = 5, a = (200 - 110) / 15 = 90 / 15 = 6, (a, o) = (6, 5)
* k = 1: o = 20, a = (200 - 440) / 15 = -240 / 15 = -16 → not allowed.

Thus only positive solution for both is:

Final Answer: (a, o) = (6, 5)

So the only possible combination is 6 apples and 5 oranges.
